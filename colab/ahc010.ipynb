```python
# 回転数の辞書
l1 = {
    0:2,
    1:1,
    2:0,
    3:3,
    4:0,
    5:1
}
```

```python
l2 = {
    0:1,
    1:0,
    2:3,
    3:2,
    4:1,
    5:0
}
```

```python
l3 = {
    0:0,
    1:3,
    2:2,
    3:1,
    4:0,
    5:1
}
```

```python
l4 = {
    0:3,
    1:2,
    2:1,
    3:0,
    4:1,
    5:0
}
```

```python
def is_l1(i, j):

  if (0 <= i <= 14) and (0 <= j <= 14) and (abs(i)==abs(j)):
    return True
  else:
    return False
  
def is_l2(i, j):

  if (15 <= i <= 29) and (0 <= j <= 14) and (abs(i)==abs(j)):
    return True
  else:
    return False

def is_l3(i, j):

  if (15 <= i <= 29) and (15 <= j <= 29) and (abs(i)==abs(j)):
    return True
  else:
    return False

def is_l4(i, j):

  if (0 <= i <= 14) and (15 <= j <= 29) and (abs(i)==abs(j)):
    return True
  else:
    return False
  
def is_curve(n:int):
  if (n != 6) and (n != 7):
    return True
  else:
    return False
```

```python
ans = ""
for j in range(30):
  s = list(input())
  for i in range(30):
    n = int(s[i])
    if is_l1(i, j) and is_curve(n):
      ans+=str(l1[n])

    elif is_l2(i, j) and is_curve(n):
      ans+=str(l2[n])

    elif is_l3(i, j) and is_curve(n):
      ans+=str(l3[n])

    elif is_l4(i, j) and is_curve(n):
      ans+=str(l4[n])
    else:
      ans+="0"

print(ans)
print(len(ans))
```
